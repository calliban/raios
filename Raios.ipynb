## 1. Reading Files

### 1.1. WRLR-raw Files processed in March 2014

Those files are in the output format for generating the WRLR functions.
They were obtained from Rainfall data (from Bauru Radar - IPMET/UNESP) and Lightning data (from RINDAT).

```python
import pandas as pd

dats = pd.read_csv('/home/joao.victor/data/wrlr_2014/Bauru_4_4_222711.txt',
                   sep='\t',
                   header=0,
                   names=('Rainfall', 'Lightning', 'Date'),
                   parse_dates=[2],
                   keep_date_col=True
                   )
```

```python
from ggplot import *

ggplot(aes(x='Lightning', y='Rainfall'), data=dats) + \
    geom_point(color='lightblue') + \
    stat_smooth(span=0.15, color='black', se=True)
```

```python
from ggplot import *

d2 = dats.groupby(dats.Date).sum()
d2['Date'] = d2.index.tolist()

ggplot(aes(x='Lightning', y='Date'), data=d2) + \
    geom_point(color='lightblue') + \
    stat_smooth(span=0.15, color='black', se=True)
    
```

```python
from sklearn.svm import SVR
from sklearn import linear_model

import numpy as np
import matplotlib.pyplot as plt

sampled_df = dats
#sampled_df = dats.sort('Rainfall')
#sampled_df = dats.sort('Lightning')
sampled_df = sampled_df[sampled_df.Date < pd.to_datetime("2009-1-2")]

X = np.array([sampled_df.Lightning]).transpose()
y = sampled_df.Rainfall.ravel()

clf = linear_model.SGDRegressor()
y_sgrd = clf.fit(X, y).predict(X)
```

```python
d3 = dats.groupby('Lightning').median()
d3['Lightning'] = d3.index.to_native_types()
from ggplot import *

ggplot(aes(x='Lightning', y='Rainfall'), data=d3) + \
    geom_point(color='lightblue') + \
    stat_smooth(span=0.15, color='black', se=True)
```

```python
plt.scatter(X, y, c='k', label='data')
plt.hold('on')
plt.plot(X, y_sgrd, c='g', label='SGDRegressor')
plt.xlabel('data')
plt.ylabel('target')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```
